## Identify the best random number

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression  # or RandomForestRegressor
from sklearn.metrics import r2_score
import pandas as pd
Train = []
Test = []
CV = []
for i in range(0, 101):
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Regression model (you can swap with RandomForestRegressor or SVR)
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predictions
    ypred_train = model.predict(X_train)
    ypred_test = model.predict(X_test)

    # RÂ² scores
    Train.append(r2_score(y_train, ypred_train))
    Test.append(r2_score(y_test, ypred_test))

    # Cross-validation RÂ²
    CV.append(cross_val_score(model, X_train, y_train, cv=5, scoring="r2").mean())

# Create dataframe
em = pd.DataFrame({"Train": Train, "Test": Test, "CV": CV})

# Filter for stable scores
gm = em[(abs(em["Train"] - em["Test"]) <= 0.05) & (abs(em["Test"] - em["CV"]) <= 0.05)]

# Get the best random_state
rs = gm[gm["CV"] == gm["CV"].max()].index.to_list()[0]
print("Best random_state number:", rs)
```

### 2.Lasso Regression Algorithm

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
# Step 1: Hyperparameter Tuning using GridSearchCV
estimator = Lasso()
param_grid = {"alpha": list(range(1, 100))}  # Trying alpha values from 1 to 99
model_hp = GridSearchCV(estimator, param_grid, cv=5, scoring="r2")

# Fit the model using GridSearchCV to find the best alpha
model_hp.fit(X_train, y_train)

# Get the best alpha
print("Best Alpha from GridSearchCV:", model_hp.best_params_)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

# Step 2: Train Lasso using the best alpha found
lasso_best = Lasso(alpha=model_hp.best_params_['alpha'])
lasso_best.fit(X_train, y_train)

# Step 3: Print Intercept and Coefficients
print("Intercept:", lasso_best.intercept_)
print("Coefficients:", lasso_best.coef_)

# Step 4: Predictions and Evaluations
ypred_train = lasso_best.predict(X_train)
print("Train R2:", r2_score(y_train, ypred_train))

print("CV Score:", cross_val_score(lasso_best, X_train, y_train, cv=5).mean())

ypred_test = lasso_best.predict(X_test)
print("Test R2:", r2_score(y_test, ypred_test))
```

### 3.Ridge Regression Algorithm

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

# Step 1: Hyperparameter Tuning using GridSearchCV
estimator=Ridge()
param_grid={"alpha":list(range(1,100))}
model_hp=GridSearchCV(estimator,param_grid,cv=5,scoring="r2")

# Fit the model using GridSearchCV to find the best alpha
model_hp.fit(X_train,y_train)

# Get the best alpha
print("Best Alpha from GridSearchCV:", model_hp.best_params_)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

# Step 2: Train Ridge using the best alpha found
ridge_best = Ridge(alpha=model_hp.best_params_['alpha'])
ridge_best.fit(X_train, y_train)

# Step 3: Print Intercept and Coefficients
print("Intercept:", ridge_best.intercept_)
print("Coefficients:", ridge_best.coef_)

# Step 4: Predictions and Evaluations
ypred_train = ridge_best.predict(X_train)
print("Train R2:", r2_score(y_train, ypred_train))

print("CV Score:", cross_val_score(ridge_best, X_train, y_train, cv=5).mean())

ypred_test = ridge_best.predict(X_test)
print("Test R2:", r2_score(y_test, ypred_test))
```

### 5.Polynomial Regression Algorithm

```python
from sklearn.preprocessing import PolynomialFeatures

# Generate polynomial features of degree 1 (which is just linear regression)
polynomial_converter = PolynomialFeatures(degree=1)
X_train_poly = pd.DataFrame(polynomial_converter.fit_transform(X_train))

from sklearn.linear_model import LinearRegression
# Train linear regression model
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Predict and evaluate on training data
ypred_train = model.predict(X_train_poly)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, ypred_train)))
print("Train R2:", r2_score(y_train, ypred_train))

from sklearn.model_selection import cross_val_score
print("Cross Validation Score:", cross_val_score(model, X_train_poly, y_train, cv=5).mean())

# Transform test data and evaluate model
X_test_poly = pd.DataFrame(polynomial_converter.fit_transform(X_test))
ypred_test = model.predict(X_test_poly)

print("Test RMSE:", np.sqrt(mean_squared_error(y_test, ypred_test)))
print("Test R2:", r2_score(y_test, ypred_test))
```

### 6.Support Vector Regression Algorithm

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, cross_val_score

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Base model
svr = SVR()

# Grid search
grid_search = GridSearchCV(svr, param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

# Best model from grid search
best_model = grid_search.best_estimator_

# Evaluation
print("Best Parameters:", grid_search.best_params_)
print("Train R2:", best_model.score(X_train, y_train))
print("Test R2:", best_model.score(X_test, y_test))

# Cross-validation score using best model
cv_score = cross_val_score(best_model, X, y, cv=5, scoring='r2')
print("Cross Validation R2 Score:", cv_score.mean())

```

### 7.Random Forest Regression Algorithm

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score

# Define the parameter grid for tuning
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize the model
rf = RandomForestRegressor(random_state=8)

# Grid search
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

# Best model from grid search
best_model = grid_search.best_estimator_

# Evaluation
print("Best Parameters:", grid_search.best_params_)
print("Train R2:", best_model.score(X_train, y_train))
print("Test R2:", best_model.score(X_test, y_test))

# Cross-validation score
cv_score = cross_val_score(best_model, X, y, cv=5, scoring='r2')
print("Cross Validation R2 Score:", cv_score.mean())
```
